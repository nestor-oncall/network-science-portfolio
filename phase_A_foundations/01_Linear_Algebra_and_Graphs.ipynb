```python
#load numpy library
import numpy as np
```

```python
#trivial examples
matA= np.matrix("1,2,4;2,3,1;-4,1,2")
vector_x= np.matrix("2;2;3")
print(matA.shape)
print(vector_x.shape)
```

```python
vector_b= np.dot(matA,vector_x)
print(vector_b)
```

```python
#a non-singular matrix (unique solution),
A=np.matrix("1,4,2;2,3,5;-1,2,1")
b=np.matrix("7;4;1")
x=np.linalg.solve(A,b)
print("Solution x =\n", x)
```

```python
# a singular matrix (infinite solutions or none)
A2=np.matrix("1,2,3;2,4,6;1,1,2")
b2=np.matrix("6;12;5")
x2=np.linalg.pinv(A2)@b2
print("One solution x =\n", x2)
print("Check Ax =\n", A2 @ x2)
```
